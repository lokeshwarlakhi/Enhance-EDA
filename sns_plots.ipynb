```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('./datasets/final_FPP.csv')
df.head()
```

```python
df.info()
```

# piechart


```python
# which flight was boarder the highest, ploting pie chart for no. of airlines
plt.figure(figsize=(10,10))
airlines_names = df.Airline.value_counts().index
airline_count = df.Airline.value_counts().values
plt.pie(airline_count,labels=airlines_names,autopct='%1.2f%%');
```

# barplot

```python
# over the year, no. of times each flight was choosen 
df.Year.unique() # the year is same, so we need to check for months
df.Month.unique() #3,4,5,6
# plotting for every month each airline's count
# air_count = 
airline_stats = df.groupby(['Airline','Month']).size().reset_index().rename(columns={0:'airline count'})
sns.barplot(x='Airline',data = airline_stats,y='airline count',hue='Month')
```

```python
airline_stats = df.groupby(['Month','Airline']).size().reset_index().rename(columns={0:'airline count'})
plt.figure(figsize=(10,10))
sns.barplot(x='Month',data = airline_stats,y='airline count',hue='Airline')
plt.legend(ncol = 3,loc = 'upper left');
```
